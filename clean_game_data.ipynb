```python
values = ['3DS' , 'DS' , 'Dreamcast' , 'GameBoyAdvance' , 'GameCube' , 'Nintendo64' , 'PSP' , 'PlayStationVita' , 'Stadia' , 'Wii' , 'WiiU', 'PlayStation', 'PlayStation2', 'Xbox', 'Xbox360']
```
